```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean = np.array([1,1]).T
```

```python
cov = np.array([
    [1,0.2],
    [0.2,2]
])
```

```python
X = np.random.multivariate_normal(mean, cov, 300)
```

```python
plt.scatter(X[:,0], X[:,1], c='blue')
plt.axis('equal')
plt.show()
```

```python
# to check it is true

sns.distplot(X, hist=True)
plt.show()
```
